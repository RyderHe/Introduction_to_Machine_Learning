```python
import pandas as pd
import numpy as np
df_X = pd.read_csv('spambase_X.csv',header=None)
df_y = pd.read_csv('spambase_y.csv',header=None)
```

```python
X = df_X.values.transpose()
y = df_y.values
```

```python
def perceptron(X, y, w, b, max_pass):
    mistake = np.repeat([0], max_pass)
    for t in range(0, max_pass):
        n = len(y)
        for i in range(0,n): 
            if (y[i][0] * (np.dot(X[i],w) + b)) <= 0:
                w = w + y[i][0] *X[i]
                b = b + y[i][0]
                mistake[t] = mistake[t] + 1
    return w,b, mistake
```

```python
zero = [0]
zero = np.repeat(zero,57)

w,b,mistakes = perceptron(X,y,zero,0,500)
```
